```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re

filepath = "./MinimisedData.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
len(df)
```

```python
df = df.dropna()
len(df)
```

```python
df = df[df.Category == df.Category[1]]
len(df)
```

```python
app_names = df[df.columns[0]]
len(app_names)
```

```python
df = df.drop([df.columns[0], df.columns[1], df.columns[2]], axis=1)
df.head()
```

```python
installs = []
for i in df.Installs:
    installs.append(int(re.sub(r"[,]","",i[:-1])))
df["install"] = installs
df.head()
```

```python
df = df.drop([ df.columns[1]],  axis=1)
df['Content Rating'].unique()
```

```python
encode = {}
k = 0
for i in df['Content Rating'].unique():
    encode[i] = k
    k+=1
encode
```

```python
content = []
for i in df['Content Rating']:
    content.append(encode[i])
df['Content'] = content
```

```python
df = df.drop([df.columns[3]],  axis=1)
df.head()
```

```python
size = []
sum = 0
num = 0
for i in df.Size:
    if i == "Varies with device":
        size.append(-1)
    else:
        size.append(float(i[:-1]))
        sum+=float(i[:-1])
        num+=1
avg = sum/num
for i in range(len(size)):
    if size[i]==-1:
        size[i] = avg
df['Size'] = size
df.head()
```

```python
df['Last Updated'][5]
```

```python
from datetime import datetime
mon = []
for i in df['Last Updated']:
    st = i.split(" ")
    t = datetime.strptime(st[0],"%B")
    k = int(t.strftime("%m"))
    mon.append(12-k+12*(2021 - int(st[-1])))
df['Last Updated'] = mon
df.head()
```

```python
ver = []
sum = 0
num = 0
for i in df['Minimum Android Version']:
    if i == "Varies with device":
        ver.append(-1)
    else:
        ver.append(float(i.split(" ")[0][0]))
        sum+=float(i.split(" ")[0][0])
        num+=1
        
avg = sum/num
for i in range(len(ver)):
    if ver[i] == -1:
        ver[i] = avg
df["Minimum Android Version"] = ver
df.head()
```

```python
ver = []
sum = 0
num = 0
for i in df['Latest Version of App']:
    if i == "Varies with device" or i.startswith('P') or i.startswith('H') or i.startswith('C') or i.startswith('L'):
        ver.append(-1)
    else:
        ver.append(float(i.split(" ")[0][0]))
        sum+=float(i.split(" ")[0][0])
        num+=1
        
avg = sum/num
for i in range(len(ver)):
    if ver[i] == -1:
        ver[i] = avg
df["Latest Version of App"] = ver
df.head()
```

```python
df.Price.unique()
```

```python
price = []
for i in df.Price:
    if i == '0':
        price .append(0.0)
    else:
        price.append(float(i[1:]))
df.Price = price
df.head()
```

```python
df["App_Names"] = app_names
df.head()
```

```python
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df.head()
```

```python
df = df.drop([df.columns[-3]],axis=1)
df.head()
```

```python
df.to_csv("./heathapps.csv")
```
