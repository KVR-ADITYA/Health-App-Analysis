```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import math

filepath = "./heathapps.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
df = df.drop([df.columns[0],df.columns[1]], axis = 1)
columns = df.columns
df.head()
```

```python
type(df['Size'][10])
```

```python
from sklearn import preprocessing
x = df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df.columns = columns
df.head()
```

```python
for i in range(len(df)):
    for j in df.columns:
        if df[j][i] == 0:
            df[j][i] = 0
        else:
            df[j][i] = df[j][i] * math.log(df[j][i])
df.head()
```

```python
entropies = []

for i in df.columns:
    temp = 0
    for j in range(len(df)):
        temp += df[i][j]/ math.log(len(df))
    entropies.append(temp)
entropies
```

```python
diver = []
for i in entropies:
    diver.append(1+i)
diver
```

```python
weights = []
for i in diver:
    weights.append(i/sum(diver))
weights
```

```python
wei = {}
for i in range(len(df.columns)):
    wei[df.columns[i]] = weights[i]
wei
```

```python
df = pd.read_csv("./normalised_data.csv")
df = df.drop([df.columns[0],df.columns[1]], axis = 1)
for i in range(len(df)):
    for j in df.columns:
        df[j][i] = df[j][i] * wei[j]
df.head()
```
